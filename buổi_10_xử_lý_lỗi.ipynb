```python
#W10A3
try:
    a = input()
    if a.isalpha():
        raise ValueError('a phải là số')
    else:
        b = int(a)
        s = set()
        while b != 1:
            if b in s:
                print('NO')
                break
            s.add(b)
            b = sum(int(ch) ** 2 for ch in str(b))
        else:
            print('YES')
except ValueError as e:
    print(e)
```

```python
#W10A7
try:
    x = input()
    if x.isalpha():
        raise TypeError('Tuoi khong hop le')
    if int(x) < 0:
        raise ValueError('Tuoi khong hop le')
except TypeError as e:
    print(e)
except ValueError as f:
    print(f)
else:
    print(2025 - int(x))
```

```python
#W10A8
file = input()
lst = []
for ch in file:
    if ch == '.':
        lst.append(ch)
        break
if len(lst) == 0:
    print('File khong hop le')
else:
    try:
        name = list(file.split('.'))
        if name[-1].lower() == 'txt' or name[-1].lower() == 'zip':
            print('Doc file thanh cong')
        else:
            raise FileExistsError('File khong hop le')
    except FileExistsError as e:
        print(e)
```

```python
#W10A10
try:
    N = int(input())
    mang = list(map(int, input().split()))
    lst = []
    tong = 0
    for i in range(N):
        if mang[i] not in lst:
            lst.append(mang[i])
            tong += mang[i]
        elif mang[i] in lst:
            raise ValueError('Mang khong hop le')
except ValueError as e:
    print(e)
else:
    print(tong)
```
