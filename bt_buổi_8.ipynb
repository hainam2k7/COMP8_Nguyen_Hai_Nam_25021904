```python
#W8A1
class Cylinder:
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height
    def surrounding_area(self):
        return (2 * self.radius * 3.14) * self.height
    def volume(self):
        return ((self.radius)**2) * 3.14 * self.height
r, h = map(float, input().split())
result = Cylinder(r, h)
print(f'{result.surrounding_area():.2f} {result.volume():.2f}')
```

```python
#W8A2
class Date:
    def __init__(self, day, month, year):
        self.day = day
        self.month = month
        self.year = year
    def is_leap_year(self):
        return (self.year % 400 == 0) or (self.year % 4 == 0 and self.year % 100 != 0)
    def is_valid(self):
        if self.year < 1 or self.month < 1 or self.month > 12 or self.day < 1:
            return False
        last_days_in_months = [31,28,31,30,31,30,31,31,30,31,30,31]
        if self.is_leap_year():
            last_days_in_months[1] = 29
        return self.day <= last_days_in_months[self.month - 1]
    def next_date(self):
        last_days_in_months = [31,28,31,30,31,30,31,31,30,31,30,31]
        if self.is_leap_year():
            last_days_in_months[1] = 29
        self.day += 1
        if self.day > last_days_in_months[self.month - 1]:
            self.day = 1
            self.month += 1
            if self.month > 12:
                self.month = 1
                self.year += 1
        return f'{self.day:02d}/{self.month:02d}/{self.year:04d}'
d, m, y = map(int, input().split('/'))
result = Date(d, m, y)
if not result.is_valid():
    print('INVALID')
else:
    print(result.next_date())
```

```python
#W8A3
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def coordinate_origin(self):
        if self.x == 0 and self.y == 0:
            return True
    def in_horizontal_axis(self):
        if self.x != 0 and self.y == 0:
            return True
    def in_vertical_axis(self):
        if self.x == 0 and self.y != 0:
            return True
    def distance(self):
        return f'{(self.x**2 + self.y**2)**0.5:.2f}'
X, Y = map(int, input().split())
result = Point(X, Y)
if result.coordinate_origin() == True:
    print('M là gốc tọa độ.')
elif result.in_horizontal_axis() == True:
    print('M thuộc trục hoành.')
elif result.in_vertical_axis() == True:
    print('M thuộc trục tung.')
else:
    print('M không thuộc trục hoành hay trục tung.')
print(result.distance())
```

```python
#W8A4
class Caculator:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def add(self): #1
        return self.x + self.y
    def subtract(self): #2
        return self.x - self.y
    def product(self): #3
        return self.x * self.y
    def divide(self): #4
        return self.x / self.y
    def power(self): #5
        return self.x**self.y
    def mod(self): #6
        return f'{self.x // self.y} dư {self.x % self.y}'
print('Menu máy tính')
caculating = True
while caculating:
    a, b = map(int, input('Chọn 2 số: ').split())
    m = input('Chọn thao tác: ')
    result = Caculator(a, b)
    if m == '1':
        print(f'a + b = {result.add()}')
    elif m == '2':
        print(f'a - b = {result.subtract()}')
    elif m == '3':
        print(f'a x b = {result.product()}')
    elif m == '4':
        print(f'a / b = {result.divide()}')
    elif m == '5':
        print(f'a ^ b = {result.power()}')
    elif m == '6':
        print(f' a / b = {result.mod()}')
    elif m == '7': #set numbers
        continue
    elif m == 'Exit':
        x = input('Bạn có muốn thoát?')
        if x == 'yes' or x == 'Yes' or x == 'YES':
            caculating = False
```

```python
#W8A5
class ShoppingCart:
    def __init__(self):
        self.products_and_price = {}
    def add(self, name, price): #2
        self.products_and_price[name] = self.products_and_price.get(name, 0) + price
    def remove(self, name): #3
        if name in self.products_and_price:
            del self.products_and_price[name]
    def is_empty(self): #4
        if not self.products_and_price:
            return True
        else:
            return False
    def total(self): #5
        return sum(self.products_and_price.values())
    def show_products(self): #6
        if self.is_empty():
            pass
        else:
            print('Danh sách sản phẩm:')
            for name, price in self.products_and_price.items():
                print(f'- {name}: {price}')
    def clear_all(self): #7
        self.products_and_price.clear()
print('____MENU____ \n' \
'1. Khởi tạo với danh sách rỗng \n' \
'2. Thêm vào giỏ hàng một sản phẩm mới \n' \
'3. Xóa khỏi giỏ một sản phẩm nào đó (tìm theo tên để xóa) \n' \
'4. Kiểm tra xem giỏ hàng có rỗng không \n' \
'5. Tính tổng tiền giỏ hàng \n' \
'6. Hiển thị toàn bộ sản phẩm trong giỏ \n' \
'7. Xóa toàn bộ giỏ hàng')
cart = ShoppingCart()
while True:
    choice = input('Nhập lệnh:')
    if choice == '2':
        name, price = input().split()
        cart.add(name, float(price))
    elif choice == '3':
        name = input()
        cart.remove(name)
    elif choice == '4':
        if cart.is_empty():
            print('Giỏ hàng rỗng.')
        else:
            print('Giỏ hàng không rỗng.')
    elif choice == '5':
        print('Tổng tiền:', cart.total())
    elif choice == '6':
        cart.show_products()
    elif choice == '7':
        cart.clear_all()
    elif choice == 'stop':
        print('Có muốn thoát khỏi chương trình không?')
        answer = input()
        if answer == 'yes':
            break
        else:
            continue
```

```python
#W8A6
class Rectangle:
    def __init__(self,w, h):
        self.w = w
        self.h = h
    def scale(self, k):
        self.w *= k
        self.h *= k
    def area(self):
        return f'{self.w * self.h:.2f}'
    def perimeter(self):
        return f'{2*(self.w + self.h):.2f}'
W, H, K = map(float, input().split())
result = Rectangle(W, H)
result.scale(K)
print(f'{result.area()} {result.perimeter()}')
```

```python
#W8A7
class fraction:
    def __init__(self, tu, mau):
        if mau == 0:
            raise ValueError('Mẫu không thể bằng 0')
        if mau < 0:
            mau = -mau
            tu = -tu
        g = self.rut_gon(abs(tu), abs(mau))
        self.tu = tu // g
        self.mau = mau // g
    def rut_gon(self, m, n):
        while n != 0:
            m, n = n, m % n
        return m
    def add(self, other):
        return fraction(self.tu * other.mau + other.tu * self.mau, self.mau * other.mau)
    def sub(self, other):
        return fraction(self.tu * other.mau - other.tu * self.mau, self.mau * other.mau)
    def mul(self, other):
        return fraction(self.tu * other.tu, self.mau * other.mau)
    def div(self, other):
        return fraction(self.tu * other.mau, self.mau * other.tu)
    def __str__(self):
        return f'{self.tu}/{self.mau}'
a, b, op, c, d = input().split()
a, b, c, d = int(a), int(b), int(c), int(d)
frac_1 = fraction(a, b)
frac_2 = fraction(c, d)
if op == '+':
    result = frac_1.add(frac_2)
elif op == '-':
    result = frac_1.sub(frac_2)
elif op == '*':
    result = frac_1.mul(frac_2)
elif op == '/':
    result = frac_1.div(frac_2)
print(result)
```

```python
#W8A8
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance
    def __str__(self):
        return f'{self.balance:.2f}'
    def deposit_or_withdraw(self, command, value):
        self.command = command
        self.val = value
        if self.command == 'DEPOSIT':
            self.balance += self.val
        elif self.command == 'WITHDRAW' and self.val <= self.balance:
            self.balance -= self.val
        elif self.command == 'WITHDRAW' and self.val > self.balance:
            pass
name, balance = input().split()
bankaccount = BankAccount(name, float(balance))
n = int(input())
for i in range(n):
    com, val = input().split()
    bankaccount.deposit_or_withdraw(com, float(val))
print(bankaccount.__str__())
```

```python
#W8A9
class Student:
    def __init__(self, name):
        self.name = name
        self.scores = {}
    def add_score(self, subject, score):
        self.scores[subject] = score
    def __str__(self):
        return f'{self.name} {self.avg():.2f} {self.rank()}'
    def avg(self):
        if len(self.scores) == 0:
            return 0
        total_score = 0
        for value in self.scores.values():
            total_score += value
        return total_score / len(self.scores)
    def rank(self):
        if self.avg() >= 8:
            return 'Excellent'
        elif self.avg() < 8 and self.avg() >= 6.5:
            return 'Good'
        elif self.avg() < 6.5 and self.avg() >= 5:
            return 'Average'
        else:
            return 'Poor'
name = input()
student = Student(name)
n = int(input())
for i in range(n):
    subject, score = input().split()
    student.add_score(subject, float(score))
print(student.__str__())
```

```python
#W8A10
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year
class Library:
    def __init__(self):
        self.author_count = {}
        self.year_count = {}
    def add(self, book):
        if book.author in self.author_count:
            self.author_count[book.author] += 1
        else:
            self.author_count[book.author] = 1
        if book.year in self.year_count:
            self.year_count[book.year] += 1
        else:
            self.year_count[book.year] = 1
    def count_by_author(self, name):
        self.name = name
        return self.author_count[self.name]
    def find_by_year(self, y):
        self.y = y
        return self.year_count[self.y]
library = Library()
n = int(input())
for i in range(n):
    line = input()
    if line.startswith('ADD'):
        data = line.split(' ', 1)
        title, author, year = data[1].split(';')
        book = Book(title, author, int(year))
        library.add(book)
    elif line.startswith('COUNTYEAR'):
        com, year = line.split()
        print(library.find_by_year(int(year)))
    elif line.startswith('COUNT'):
        com, author = line.split()
        print(library.count_by_author(author))
```

```python
#W8A11
import math
class Point:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z
    def vector_to(self, other):
        return Vector(
            other.x - self.x,
            other.y - self.y,
            other.z - self.z
        )

class Vector:
    def __init__(self,x ,y, z):
        self.x = x
        self.y = y
        self.z = z
    def dot(self, other):
        return self.x * other.x + self.y * other.y + self.z * other.z
    def cross(self, other):
        return Vector(
            self.y * other.z - self.z * other.y,
            self.z * other.x - self.x * other.z,
            self.x * other.y - self.y * other.x
        )
    def norm(self):
        return self.dot(self)**0.5

class Angle:
    def __init__(self, A, B, C, D):
        self.A = A
        self.B = B
        self.C = C
        self.D = D
    def angle(self):
        AB = A.vector_to(B)
        AC = A.vector_to(C)
        BC = B.vector_to(C)
        BD = B.vector_to(D)

        n1 = AB.cross(AC)
        n2 = BC.cross(BD)
        cos_phi = abs(n1.dot(n2)/(n1.norm() * n2.norm()))
        return math.acos(cos_phi) * 180/3.14

A = Point(*map(float, input().split()))
B = Point(*map(float, input().split()))
C = Point(*map(float, input().split()))
D = Point(*map(float, input().split()))
phi = Angle(A, B, C, D).angle()
print(f'{phi:.2f}')
```

```python
#W8A12
class Complex:
    def __init__(self, r, i):
        self.r = r
        self.i = i
    def complex_nums(self, other):
        return Complex(self.r * other.r - self.i * other.i, self.r * other.i + self.i * other.r)
    def __str__(self):
        return f'{self.r} {self.i}'
r_1, i_1 = map(int, input().split())
r_2, i_2 = map(int, input().split())
complex_1 = Complex(r_1, i_1)
complex_2 = Complex(r_2, i_2)
result = complex_1.complex_nums(complex_2)
print(result.__str__())
```

```python
#W8A13
class Triangle:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
    def is_valid(self):
        if self.a + self.b > self.c and self.a + self.c > self.b and self.b + self.c > self.a:
            return True
        else:
            return False
    def area(self):
        p = (self.a + self.b + self.c)/2
        S = (p*(p - self.a)*(p - self.b)*(p - self.c))**0.5
        return f'{S:.2f}'
a, b, c = map(int, input().split())
triangle = Triangle(a, b, c)
if not triangle.is_valid():
    print('invalid')
else:
    print(triangle.area())
```
