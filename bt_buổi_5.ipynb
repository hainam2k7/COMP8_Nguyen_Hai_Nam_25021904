```python
#W5A1
for i in range(5):
  a, b = map(int,input().split())
  def Max_of_Two(a, b):
    return max(a, b)
  print(Max_of_Two(a, b))
```

```python
#W5A2
for i in range(5):
  a, b = map(int,input().split())
  number = (a,b)
  def swap(a,b):
    return number[::-1]
  print(swap(a,b))
```

```python
#W5A3
for i in range(5):
  a = int(input())
  def số_nguyên_tố(a):
    if a >= 2:
        for x in range(2, int(a**0.5) + 1):
            if a % x == 0:
                return 'False'
        return 'True'
    else:
        return 'False'
  print(số_nguyên_tố(a))
```

```python
#W5A4
for x in range(5):
  a = int(input())
  def số_hoàn_hảo(a):
    ước = 0
    for i in range(1, a):
        if a%i == 0:
            ước += i
    if ước == a:
        return 'True'
    else:
        return 'False'
  print(số_hoàn_hảo(a))
```

```python
#W5A5
for x in range(5):
  số_nguyên = input().split()
  k = int(input())
  danh_sách = [số_nguyên]
  def tìm_k(số_nguyên, k):
    for i in range(len(danh_sách[0])):
        if int(danh_sách[0][i]) == k:
            return i + 1
    return -1
  print(tìm_k(số_nguyên, k))
```

```python
#W5A6
for x in range(5):
  a = int(input())
  def giai_thừa(a):
    if a == 0 or a == 1:
        return 1
    elif a >= 2:
        kết_quả = 1
        for i in range(1, a + 1):
            kết_quả *= i
        return kết_quả
    else:
        return "Không thể tính giai thừa của số âm"
  print(giai_thừa(a))
```

```python
#W5A7
for x in range(5):
  a = input()
  phép_toán = [a.split()]
  def operation(phép_toán):
    if phép_toán[0][1] == '+':
        return f'{float(phép_toán[0][0]) + float(phép_toán[0][2]):.2f}'
    elif phép_toán[0][1] == '-':
        return f'{float(phép_toán[0][0]) - float(phép_toán[0][2]):.2f}'
    elif phép_toán[0][1] == '*':
        return f'{float(phép_toán[0][0]) * float(phép_toán[0][2]):.2f}'
    elif phép_toán[0][1] == '/':
        return f'{float(phép_toán[0][0]) / float(phép_toán[0][2]):.2f}'
  print(operation(phép_toán))
```

```python
#W5A8
for x in range(5):
  x, y = map(int, input().split())
  def hamming_distance(x, y):
    xor = x ^ y
    count = 0
    while xor > 0:
        count += xor & 1
        xor >>= 1
    return count
  print(hamming_distance(x, y))
```

```python
#W5A9
for x in range(5):
  n = input()
  def tổng_chữ_số(n):
    tổng = 0
    for i in range(0, len(n)):
        tổng += int(n[i])
    return tổng
  print(tổng_chữ_số(n))
```

```python
#W5A10
for x in range(5):
  a, b = input().split()
  def đẳng_cấu(a, b):
    if len(a) != len(b):
        return 'False'
    else:
        for i in range(len(a)):
            if a.count(a[i]) != b.count(b[i]):
                return 'False'
        return 'True'
  print(đẳng_cấu(a, b))
```
