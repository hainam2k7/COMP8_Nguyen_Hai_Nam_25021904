```python
#W6A1
numbers = input().split()
number_list = [int(num) for num in numbers]
def eliminate_duplicates(num_list):
    result = []
    for i in range(0, len(num_list)):
        if num_list[i] not in result:
            result.append(num_list[i])
        else:
            continue
    return result
print(eliminate_duplicates(number_list))
```

```python
#W6A2
numbers = input().split()
number_list = [int(num) for num in numbers]
def tong_tich_luy(num_list):
    total = [num_list[0]]
    for i in range(1, len(num_list)):
        num_list[i] += num_list[i - 1]
        total.append(num_list[i])
    return total
print(tong_tich_luy(number_list))
```

```python
#W6A3
numbers = tuple(map(int,input().split()))
k = int(input())
nums_list = list(numbers)
def change_position(num_tuple, k):
    while k != 0:
        nums_list.append(nums_list[0])
        nums_list.pop(0)
        k -= 1
    return tuple(num_tuple)
print(change_position(nums_list, k))
```

```python
#W6A5
numbers = list(map(int,input().split()))
def counting(num):
    positives = 0
    negatives = 0
    zeros = 0
    for i in range(len(num)):
        if num[i] > 0:
            positives += 1
        elif num[i] < 0:
            negatives += 1
        else:
            zeros += 1
    return positives, negatives, zeros
result = list(counting(numbers))
dictionary = {'positives':result[0],'negatives':result[1],'zeros':result[2]}
print(dictionary)
```

```python
#W6A9
dict_1 = {}
for p in input().split():
    key, value = p.split(':')
    dict_1[key] = int(value)
dict_2 = {}
for p in input().split():
    key, value = p.split(':')
    dict_2[key] = int(value)
merged = {}
for key in dict_1:
    merged[key] = dict_1[key]
for key in dict_2:
    if key in merged:
        merged[key] += dict_2[key]
    else:
        merged[key] = dict_2[key]
merged_sorted = dict(sorted(merged.items()))
print(merged_sorted)
```

```python
#W6A10
numbers = list(map(int, input().split()))
k = int(input())
def tong_bang_k(num, n):
    pairs = set()
    for i in range(len(num)):
        for j in range(i + 1, len(num)):
            if num[i] + num[j] == n:
                pairs.add((num[i], num[j]))
    result = sorted(pairs,key=lambda x: x[0])
    return result
print(tong_bang_k(numbers, k))
```

```python
#W6A11
numbers = tuple(map(int,input().split()))
def so_chan(num):
    chan = []
    for i in range(len(num)):
        if num[i] % 2 == 0:
            chan.append(num[i])
        else:
            continue
    return tuple(chan)
def so_le(num):
    le = []
    for i in range(len(num)):
        if num[i] % 2 == 1:
            le.append(num[i])
        else:
            continue
    return tuple(le)
print(so_chan(numbers))
print(so_le(numbers))
```

```python
#W6A14
list_1 = list(map(int, input().split()))
list_2 = list(map(int, input().split()))
def similarities(lst_1, lst_2):
    result = []
    for i in range(len(lst_1)):
        if lst_1[i] in lst_2 and lst_1[i] not in result:
            result.append(lst_1[i])
        else:
            continue
    return result
print(similarities(list_1, list_2))
```

```python
#W6A20
numbers = list(map(int, input().split()))
k = int(input())
def find_num(num, a):
    result = -1
    for i in range(len(num)):
        if num[i] == a:
            result = i + 1
            break
    return result
print(find_num(numbers, k))
```
