```python
#W9A3
def minimum_swap(num):
    digits = list(str(num))
    def helper(i, j, best):
        if i == len(digits) - 1:
            return best
        if j == len(digits):
            return helper(i + 1, i + 2, best)
        digits[i], digits[j] = digits[j], digits[i]
        value = int(''.join(digits))
        best = min(best, value)
        digits[i], digits[j] = digits[j], digits[i]
        return helper(i, j + 1, best)
    return helper(0, 1, float('inf'))
num = int(input())
print(minimum_swap(num))
```

```python
#W9A4
def match_from_start(str_1, str_2):
    if not str_1:
        return False
    if not str_2:
        return True
    if str_1[0] == str_2[0]:
        return match_from_start(str_1[1:], str_2[1:])
    else:
        return False
def bien_doi_xau(str_1, str_2):
    if not str_1:
        return False
    if not str_2:
        return True
    if match_from_start(str_1, str_2):
        return True
    return bien_doi_xau(str_1[1:], str_2)
Q = int(input())
for i in range(Q):
    a = input().lower()
    b = input().lower()
    if len(a) < len(b):
        print('NO')
    elif bien_doi_xau(a, b):
        print('YES')
    else:
        print('NO')
```

```python
#W9A6
def next_permutation(a):
    def find_i(idx):
        if idx < 0:
            return -1
        if a[idx] < a[idx + 1]:
            return idx
        return find_i(idx - 1)
    def find_j(j, idx):
        if a[idx] > a[j]:
            return idx
        return find_j(j, idx - 1)
    def reverse(l, r):
        if l >= r:
            return
        a[l], a[r] = a[r], a[l]
        reverse(l + 1, r - 1)
    i = find_i(len(a) -2)
    if i == -1:
        reverse(0, len(a) - 1)
        return a
    j = find_j(i, len(a) -1)
    a[i], a[j] = a[j], a[i]
    reverse(i + 1, len(a) - 1)
    return a
n = int(input())
a = list(map(int, input().split()))
res = next_permutation(a)
for x in res:
    print(x, end=' ')
```

```python
#W9A7
def CanPlaceFlowers(flowerbed, k):
    def backtrack(i, k):
        if k == 0:
            return True
        if i >= len(flowerbed):
            return False
        left = (i == 0) or (flowerbed[i - 1] == 0)
        right = (i == len(flowerbed) - 1) or (flowerbed[i + 1] == 0)
        if flowerbed[i] == 0 and left and right:
            flowerbed[i] = 1
            if backtrack(i + 2, k - 1):
                return True
        return backtrack(i + 1, k)
    return backtrack(0, k)
flowerbed = list(map(int, input().split()))
k = int(input())
if CanPlaceFlowers(flowerbed, k):
    print(True)
else:
    print(False)
```

```python
#W9A8
def is_path_crossing(moves):
    def moving(i, x, y, visited):
        if i == len(moves):
            return False
        if moves[i] == 'R':
            x += 1
        elif moves[i] == 'L':
            x -= 1
        elif moves[i] == 'U':
            y += 1
        elif moves[i] == 'D':
            y -= 1
        if (x, y) in visited:
            return True
        visited.add((x, y))
        return moving(i + 1, x, y, visited)
    visited = {(0, 0)}
    return moving(0, 0, 0, visited)
moves = input()
print(is_path_crossing(moves))
```

```python
#W9A9
def Median(nums):
    a, b, c, d, e = nums
    if a > b:
        a, b = b, a
    if c > d:
        c, d = d, c
    if b > d:
        max4 = b
        mid1 = d
    else:
        max4 = d
        mid1 = b
    if a < c:
        min4 = a
        mid2 = c
    else:
        min4 = c
        mid2 = a
    x, y = mid1, mid2
    if x > y:
        x, y = y, x
    if e < x:
        return x
    elif e > x:
        return y
    else:
        return e
nums = list(map(int, input().split()))
print(Median(nums))
```

```python
#W9A10
def findRadius(houses, heaters):
    houses.sort()
    heaters.sort()
    n, m = len(houses), len(heaters)
    def distance(i, j):
        if i == n:
            return 0
        while j < m - 1 and abs(heaters[j + 1] - houses[i]) <= abs(heaters[j] - houses[i]):
            j += 1
        d = abs(heaters[j] - houses[i])
        return max(d, distance(i + 1, j))
    return distance(0, 0)
houses = list(map(int, input().split()))
heaters = list(map(int, input().split()))
print(findRadius(houses, heaters))
```
