```python
#W7A1
lst = list(map(int, input().split()))
n = int(input())
def binary_search(arr, target):
    left, right = 0, len(arr) -1
    while left <= right:
        mid = (left + right)//2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
print(binary_search(lst, n))
```

```python
#W7A3 (Dùng Bubble Sort)
lst = list(map(int, input().split()))
def Bubble_Sort(arr):
    for i in range(len(arr)):
        swapped = False
        for j in range(0, len(arr) - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr
print(Bubble_Sort(lst))
```

```python
#W7A3 (Dùng Selection Sort)
lst = list(map(int, input().split()))
def Selection_Sort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i+1,len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr
print(Selection_Sort(lst))
```

```python
#W7A4
lst = list(map(int, input().split()))
def xuat_hien_nhieu_nhat(arr):
    dictionary = {}
    for i in arr:
        if i in dictionary:
            dictionary[i] += 1
        else:
            dictionary[i] = 1
    key_result = []
    value_result = []
    for key in dictionary:
        key_result.append(key)
        value_result.append(dictionary[key])
    for j in range(len(value_result)):
        if value_result[j] == max(value_result):
            return key_result[j], value_result[j]
result = xuat_hien_nhieu_nhat(lst)
print(f'{result[0]} xuat hien nhieu nhat, som nhat, {result[1]} lan')
```

```python
#W7A5
lst = list(map(int, input().split()))
X = int(input())
def counting(arr, n):
    result = 0
    for i in range(len(arr) - 1):
        for j in range(i + 1, len(arr)):
            if arr[i] + arr[j] == n:
                result += 1
    return result
print(counting(lst, X))
```

```python
#W7A7
lst = list(map(int, input().split()))
def length(nums):
    if len(nums) == 0:
        return 0
    dp = [1]* len(nums)
    for i in range(len(nums)):
        for j in range(i):
            if nums[j] < nums[i]:
                dp[i] = max(dp[i], dp[j] + 1)
    return max(dp)
print(length(lst))
```

```python
#W7A8
interval = []
while True:
    line = input()
    if line == '':
        break
    sublist = list(map(int, line.split()))
    interval.append(sublist)
queries = list(map(int, input().split()))
def min_intervals(interval, queries):
    result = []
    for i in range(len(queries)):
        min_val = 0
        for j in range(len(interval)):
            left, right = interval[j][0], interval[j][1]
            if left <= queries[i] and right >= queries[i]:
                val = right - left + 1
                if min_val == 0:
                    min_val += val
                else:
                    min_val = min(min_val, val)
        if min_val == 0:
            result.append(-1)
        else:
            result.append(min_val)
    return result
print(min_intervals(interval, queries))
```

```python
#W7A9
nums = list(map(int, input().split()))
def dem_so(arr):
    result = []
    for i in range(len(arr)):
        if i == len(arr) - 1:
            result.append(0)
        else:
            smaller_nums = 0
            for j in range(i + 1, len(arr)):
                if arr[j] < arr[i]:
                    smaller_nums += 1
            result.append(smaller_nums)
    return result
print(dem_so(nums))
```

```python
#W7A10
nums = list(map(int, input().split()))
m = int(input())
def maximum_subarray_sum_module(arr, k):
    prefix = 0
    max_sum = 0
    prefix_list = []
    for x in arr:
        prefix = (prefix + x) % k
        max_sum = max(max_sum, prefix)
        left, right = 0, len(prefix_list)
        while left < right:
            mid = (left + right) // 2
            if prefix_list[mid] <= prefix:
                left = mid + 1
            else:
                right = mid
        if left < len(prefix_list):
            candidate = (prefix - prefix_list[left] + k) % k
            max_sum = max(max_sum, candidate)
        left, right = 0, len(prefix_list)
        while left < right:
            mid = (left + right) // 2
            if prefix_list[mid] < prefix:
                left = mid + 1
            else:
                right = mid
        prefix_list.insert(left, prefix)
    return max_sum
print(maximum_subarray_sum_module(nums, m))
```

```python
#W7A11
sentences = list(map(str, input().split('.')))
word = input()
def counting_word(arr, value):
    result = 0
    for i in range(len(arr)):
        if word in sentences[i]:
            result += sentences[i].count(value)
    return result
print(counting_word(sentences, word))
```

```python
#W7A12
data = []
while True:
    line = input().strip()
    if line == "":
        break
    data.append(line)
is_number = True
for x in data:
    if not x.lstrip('-').isdigit():
        is_number = False
        break
if is_number:
    sorted_data = sorted(data, key=lambda x: int(x), reverse=True)
else:
    sorted_data = sorted(data)
rank = {}
i = 0
for value in sorted_data:
    rank[value] = chr(ord('A') + i)
    i += 1
result = []
for x in data:
    result.append(rank[x])
print(' '.join(result))
```

```python
#W7A13
lst = list(map(int, input().split()))
def day_hai_hoa(arr):
    dictionary = {}
    for x in arr:
        if x in dictionary:
            dictionary[x] += 1
        else:
            dictionary[x] = 1
    max_len = 0
    for x in dictionary:
        if x + 1 in dictionary:
            max_len = max(max_len, dictionary[x] + dictionary[x + 1])
    return max_len
print(day_hai_hoa(lst))
```

```python
#W7A14
n = int(input())
nums = list(map(int, input().split()))
def linear_search(arr, k):
    result = []
    for i in range(k):
        if arr[i] == 7:
            result.append(i)
    if len(result) == 0:
        return 'Not found'
    else:
        return result[::-1]
r = linear_search(nums, n)
if r == 'Not found':
    print(r)
else:
    for x in range(len(r)):
        print(r[x],end=' ')
```
